```python
import json

# Path to your JSON file
json_file_path = "./../read_csv_pdfs/output_data.json"

texts = []
with open(json_file_path, "r") as file:
    line_number = 0
    for line in file:
        line_number += 1
        try:
            # Parse each line as a separate JSON object
            json_object = json.loads(line)
            # Check if 'pdf_text' exists and is not None before appending
            if json_object and "pdf_text" in json_object:
                if json_object["pdf_text"]:
                    texts.append(json_object["pdf_text"])
                else:
                    print(f"Line {line_number}: 'pdf_text' is empty or None.")
            else:
                print(f"Line {line_number}: 'pdf_text' key not found in JSON object.")
        except json.JSONDecodeError as e:
            print(f"Line {line_number}: Error parsing JSON - {e}")

# Check if texts is not empty before proceeding to topic modeling
if texts:
    from bertopic import BERTopic

    topic_model = BERTopic()
    topics, probabilities = topic_model.fit_transform(texts)

    # Analyzing and visualizing topics as before
else:
    print("No valid text data found for topic modeling.")
```

```python
import matplotlib.pyplot as plt

# Visualize the topics
topic_vis = topic_model.visualize_topics()

# Show the plot
plt.show(topic_vis)
```
